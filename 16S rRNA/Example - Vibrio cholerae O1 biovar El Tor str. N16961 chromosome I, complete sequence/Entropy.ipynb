# Entropy 분석을 위한 그래프 그리기

아래 예시는 Vibrio cholerae O1 biovar El Tor str. N16961 chromosome I에 존재하는 8개의 16S rRNA를 FASTA 파일로 얻어 Multiple Sequence Alignment by CLUSTALW를 진행한 뒤 BioEdit 7.2를 이용해 Entropy를 계산한 결과를 entropy.txt 파일로 다운받아 진행하였다.

- Vibrio cholerae O1 biovar El Tor str. N16961 chromosome I : https://www.ncbi.nlm.nih.gov/nuccore/NC_002505.1?report=gbwithparts&log$=seqview
- Multiple Sequence Alignment by CLUSTALW : https://www.genome.jp/tools-bin/clustalw
- BioEdit 7.2 : https://bioedit.software.informer.com/7.2/

## Step By Step

```python
# entropy.txt 불러오기
f = open("entropy.txt", 'r')
data = f.read()
print(data)
f.close()
```

```python
# 필요한 부분만 잘라내기
for i in range(len(str(data))) : 
    if data[i] == data[i-1] == '\n' : # 145, 146번째인 '\n\n' 이후로 'Position 1'이 온다.
        print(i)
        entropy = data[i+1:] # '\n\n' 이후의 값들
```

```python
print(entropy)
```

```python
# First position
print(entropy.split('\n')[0].split(': ')[0]) # 'Position'이란 말은 의미가 없다.
print(entropy.split('\n')[0].split(': ')[1]) # 0.34883와 같은 Score만 의미 있는 값이다.
```

```python
# Last position
print(len(entropy.split('\n'))-1) # 'Position'이란 말을 쓰지 않기 위해 1부터 1555까지의 수를 새로 만들 것이다.
print(entropy.split('\n')[1554].split(': ')[1])
```

```python
# Score만 가져와서 %로 환산하기
entropy_score = []
for i in range(len(entropy.split('\n'))-1) : 
    score = entropy.split('\n')[i].split(': ')[1]
    per_score = float(score)*100
    entropy_score.append(per_score)

print(len(entropy_score))
print(entropy_score)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = list(range(1, len(entropy_score)+1, 1)) # Position을 의미하는 1부터 1555까지의 수
y = entropy_score

plt.figure(figsize=(40, 6))
plt.title("Entropy Plot", fontsize=60)
plt.xlabel('Position', fontsize=45)
plt.ylabel('% Entropy', fontsize=45)
plt.tick_params(axis='x', labelsize=25)
plt.tick_params(axis='y', labelsize=25)
plt.xlim(-20, len(entropy.split('\n'))-1+20)
plt.ylim(min(entropy_score), max(entropy_score)+10)
plt.bar(x, y, width=2.5)
plt.show()
```

## 함수 만들기

```python
def Entropy_plot(inp_entropy) : # Entropy_plot('Entropy_파일명.txt')
    
    import matplotlib.pyplot as plt
    %matplotlib inline

    # 파일 불러오기
    f = open(inp_entropy, 'r')
    data = f.read()
    f.close()
    
    # 필요한 부분만 잘라내기
    for i in range(len(str(data))) : 
        if data[i] == data[i-1] == '\n' :
            entropy = data[i+1:]
    
    # Score만 가져와서 %로 환산하기
    entropy_score = []
    for i in range(len(entropy.split('\n'))-1) : 
        score = entropy.split('\n')[i].split(': ')[1]
        per_score = float(score)*100
        entropy_score.append(per_score)
    
    # Plot
    x = list(range(1, len(entropy_score)+1, 1))
    y = entropy_score

    plt.figure(figsize=(40, 6))
    plt.title("Entropy Plot", fontsize=60)
    plt.xlabel('Position', fontsize=45)
    plt.ylabel('% Entropy', fontsize=45)
    plt.tick_params(axis='x', labelsize=25)
    plt.tick_params(axis='y', labelsize=25)
    plt.xlim(-20, len(entropy.split('\n'))-1+20)
    plt.ylim(min(entropy_score), max(entropy_score)+10)
    plt.bar(x, y, width=2.5)
    plt.show()
```

```python
Entropy_plot('entropy.txt')
```
