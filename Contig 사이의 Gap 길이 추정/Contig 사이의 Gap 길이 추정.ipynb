```python
def scaff_lenth(contig_num) : 
    contig = contig_num
    
    # $ bp_seq_length.pl Contig*.fa > Contig_bp.txt
    f = open("./File/CAP3/bl21_de3.cap.contigs.links", 'r')
    cont_name = contig.split(' ')[0]+str(int(contig.split(' ')[1]))
    lines = f.readlines()
    cont_list = []
    for i in  range(len(lines)) : 
        line = lines[i].split(' ')
        if '\n' not in line[0] : 
            if cont_name in line : 
                for q in range(0, 100, 2) : 
                    if int(lines[i-q].split(' ')[0][6:])+1 == int(lines[i-q].split(' ')[2][6:]) : 
                        if lines[i-q].split(' ')[0] == lines[i-q-2].split(' ')[2] : 
                            cont_list.append(lines[i-q].split(' ')[0])
                        else : 
                            cont_list.append(lines[i-q].split(' ')[0])
                            break
                break
    cont_list = cont_list[1:]       
    cont_list.reverse()

    f = open("./File/CAP3/bl21_de3.cap.contigs.links", 'r')
    cont_name = contig.split(' ')[0]+str(int(contig.split(' ')[1]))
    lines = f.readlines()
    for i in  range(len(lines)) : 
        line = lines[i].split(' ')
        if '\n' not in line[0] : 
            if cont_name in line : 
                for q in range(0, 100, 2) : 
                    if int(lines[i+q].split(' ')[0][6:])+1 == int(lines[i+q].split(' ')[2][6:]) : 
                        if lines[i+q].split(' ')[2] == lines[i+q+2].split(' ')[0] : 
                            cont_list.append(lines[i+q].split(' ')[0])
                        elif lines[i+q].split(' ')[2] != lines[i+q+2].split(' ')[0] : 
                            cont_list.append(lines[i+q].split(' ')[0])
                            cont_list.append(lines[i+q].split(' ')[2])
                            break
                break
    f.close()
    
    f = open("./File/CAP3/Contigs/Contig_bp.txt", 'r')
    cont_name = contig.split(' ')[0]+str(int(contig.split(' ')[1]))
    lines = f.readlines()
    scaff_lenth = 0
    for i in  range(len(lines)) : 
        line = lines[i].split(' ')
        if line[0][:-3] in cont_list : 
            print(line[0], ':', line[1], 'bp')
            scaff_lenth = scaff_lenth + int(line[1])

    f.close()
    print()
    print('Scaffold의 길이는 약', scaff_lenth, 'bp이다.')
```

```python
def scaffolding(contig_number, sample_name):
    f = open("./File/CAP3/Contigs/Contig_bp.txt", 'r')
    
    cont1_name = str(contig_number)

    lines = f.readlines()
    for line in lines:
        name = str(line.split(' ')[0][:-3])
        if name == cont1_name:
            con1_bp = int(line.split(' ')[1])
            print(contig_number, ':', con1_bp, 'bp')

    f.close()

    list_contig = []
    load = './File/'+contig_number+'.txt'
    f = open(load, 'r')
    lines = f.readlines()
    for line in lines:
        line = line.split(' ')
        if line[0].split('\t')[0][0] == 'F':
            name = line[0].split('\t')[0]
            list_contig.append(name)
    f.close()
    
    print('Contig'+'\t\t'+'Sample'+'\t'+'\t'+'Start'+'\t'+'End')
    load = './File/'+contig_number+'.txt'
    f = open(load, 'r')
    Contig = list_contig
    name = str(sample_name)
    lines = f.readlines()
    k = 0
    for line in lines:
        line = line.split(' ')
        if line[0].split('\t')[0][:-3] == name:
            name = line[0].split('\t')[0]
            start = int(line[0].split('\t')[1])
            con1_bp_start = int(start)
            length = int(line[0].split('\t')[2][:-1])
            end = start+length-1
            print(str(contig_number)+'\t'+str(name)+'\t'+str(start)+'\t'+str(end))
            k = k + 1
    f.close()
    
    contig_number = str(int(contig_number[6:])+1)
    load = './File/'+'Contig'+contig_number+'.txt'
    f = open(load, 'r')
    Contig = list_contig
    name = str(sample_name)
    lines = f.readlines()
    k = 0
    for line in lines:
        line = line.split(' ')
        if line[0].split('\t')[0][:-3] == name:
            name = line[0].split('\t')[0]
            start = int(line[0].split('\t')[1])
            length = int(line[0].split('\t')[2][:-1])
            end = start+length-1
            con2_bp_end = int(end)
            print('Contig'+contig_number+'\t'+str(name)+'\t'+str(start)+'\t'+str(end))
            k = k + 1
    f.close()

    result = int(con1_bp - con1_bp_start + con2_bp_end)
    print(con1_bp, '-', con1_bp_start, '+', con2_bp_end, '=', result)
    print()
    
    return result
```

```python
def scaffolding_calc(contig_number) : 
    print('-', contig_number)
    print('Sample'+'\t'+'\t'+'Start'+'\t'+'End')
    load = './File/'+contig_number+'.txt'
    f = open(load, 'r')
    Contig = read_name
    lines = f.readlines()
    k = 0
    contig1 = []
    for line in lines:
        line = line.split(' ')
        if line[0].split('\t')[0][0] == 'F':
            name = line[0].split('\t')[0]
            for q in contig:
                if q[:-1] == name:
                    name = q
            start = int(line[0].split('\t')[1])
            length = int(line[0].split('\t')[2][:-1])
            end = start+length-1
            print(str(name)+'\t'+str(start)+'\t'+str(end))
            contig1.append(str(name))
            k = k + 1
    f.close()
    print()
    
    contig_num = str(int(contig_number[6:])+1)
    print('- Contig'+contig_num)
    print('Sample'+'\t'+'\t'+'Start'+'\t'+'End')
    load = './File/'+'Contig'+contig_num+'.txt'
    f = open(load, 'r')
    Contig = read_name
    lines = f.readlines()
    k = 0
    contig2 = []
    for line in lines:
        line = line.split(' ')
        if line[0].split('\t')[0][0] == 'F':
            name = line[0].split('\t')[0]
            for q in contig:
                if q[:-1] == name:
                    name = q
            start = int(line[0].split('\t')[1])
            length = int(line[0].split('\t')[2][:-1])
            end = start+length-1
            print(str(name)+'\t'+str(start)+'\t'+str(end))
            contig2.append(str(name))
            k = k + 1
    f.close()
    print()

    print('- 일치하는 Samples')
    samecont = []
    k = 0
    for i in contig1 :
        ii = i[:-4]
        for j in contig2 : 
            if ii in j[:-4] :
                print(str(i)+'\t'+str(j))
                samecont.append(i[:-3])
                k = k + 1
    print('총', k, '개 일치하는 것을 알 수 있다.')
    print()
    
    list_mean = []
    for i in samecont : 
        list_mean.append(scaffolding(contig_number, i))
        
    import numpy as np
    mean = np.mean(list_mean)
    print('평균 :', mean) # 두 스캐폴드에서 GAP을 제외한 거리

    # $less -S anc.escherichia_coli_bl21_de3_.001
    size = 8143 # Fragment Size
    print('SIZE - 평균 =', int(size - mean))

    print('약', int(size - mean), '정도의 Size가 Gap이라고 할 수 있다.')
```
